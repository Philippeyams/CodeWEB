```python
#Entree des variables

S=float(input("Entrée la superficie du bassin versant en m2:"))
I=float(input("Entrée l'intensité de la pluie en mm:"))
alpha=float(input("Entrée le coefficient d'infiltration:"))
  
def  debit (S, I, alpha):
    return S*I*alpha

```

```python

debit_value= []
alpha_value=[]

for alpha in np.arange(0, 1, 0.01):
    #print(debit (100, 34, alpha))

    debit_val= debit (100, 34, alpha)
    
    alpha_value.append(alpha)  # Stocke alpha
    debit_value.append(debit_val) 


plt.plot(alpha_value, debit_value)

plt.xlabel('Alpha')
plt.ylabel('Débit')
plt.title('Débit en fonction de Alpha')
plt.grid(True)
plt.show()
```
